```python
import matplotlib.pyplot as plt
import numpy as np
import random

k = random.uniform(0, 1)
n = np.linspace(1, 10, 10)

def x(n):
  return np.cos(2*np.pi*k*n)

def set_functions(n, k):
  y = [0, 0, 0, 0]
  y[0] = (k) * (x(n+1) - x(n))
  y[1] = (k/2) * (x(n+1) - x(n-1))
  y[2] = (k**2) * (x(n+1) - 2*x(n) + x(n-1))
  y[3] = 0.1*x(n-2) + 0.2*x(n-1) + 0.4*x(n) + 0.2*x(n+1) + 0.1*x(n+2)
  return y

fig , axes1 = plt.subplots(1, 4, figsize=(25, 10))
y = set_functions(n, k)
titles = ['A', 'B', 'C', 'D']

for i in range (0, 4):
  axes1[i].title.set_text(titles[i])
  axes1[i].set_xlabel('n')
  axes1[i].set_ylabel('y[n]')
  axes1[i].stem(n, y[i], use_line_collection=True)
```

```python
n = np.linspace(1, 200, 200)
y = set_functions(n, k)
fig , axes2 = plt.subplots(1, 4, figsize=(50, 10))
for i in range (0, 4):
  axes2[i].title.set_text(titles[i])
  axes2[i].set_xlabel('n')
  axes2[i].set_ylabel('y[n]')
  axes2[i].stem(n, y[i], use_line_collection=True)
```
