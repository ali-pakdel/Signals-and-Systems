a)

Impulse

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal

n = np.arange(0,  100, 1)

def impulseA(n):
  return (1/2) * (signal.unit_impulse(100)[n] + signal.unit_impulse(100)[n-1])

def impulseB(n):
  return signal.unit_impulse(120)[n+1] - signal.unit_impulse(120)[n]

def impulseC(n):
  return signal.unit_impulse(120)[n+1] - 2*signal.unit_impulse(120)[n] + signal.unit_impulse(100)[n-1]

def impulseD(n):
  return np.sum([(np.cos((np.pi*m)/10)*(signal.unit_impulse(100)[n-m])) for m in range(0, 20)], axis=0)

impulse = [impulseA(n), impulseB(n), impulseC(n), impulseD(n)]
fig , axes1 = plt.subplots(1, 4, figsize=(25, 5))
titles = ['A', 'B', 'C', 'D']

for i in range (0, 4):
  axes1[i].title.set_text(titles[i])
  axes1[i].stem(impulse[i], use_line_collection=True)
```

b)

Step

```python
def set_steps(n, mu):  
  step = [np.sum([impulseA(n-m) for m in range(0, mu+1)], axis=0),
            np.sum([impulseB(n-m) for m in range(0, mu+1)], axis=0),
            np.sum([impulseC(n-m) for m in range(0, mu+1)], axis=0),
            np.sum([impulseD(n-m) for m in range(0, mu+1)], axis=0)]
  return step

step5 = set_steps(n, 5)
step15 = set_steps(n, 15)
step20 = set_steps(n, 20)
            
fig , axes2 = plt.subplots(1, 4, figsize=(20, 5))
fig.suptitle('mu = 5')

for i in range (0, 4):
  axes2[i].title.set_text(titles[i])
  axes2[i].stem(step5[i], use_line_collection=True)

fig , axes3 = plt.subplots(1, 4, figsize=(20, 5))
fig.suptitle('mu = 15')

for i in range (0, 4):
  axes3[i].title.set_text(titles[i])
  axes3[i].stem(step15[i], use_line_collection=True)

fig , axes4 = plt.subplots(1, 4, figsize=(20, 5))
fig.suptitle('mu = 20')

for i in range (0, 4):
  axes4[i].title.set_text(titles[i])
  axes4[i].stem(step20[i], use_line_collection=True)
```
