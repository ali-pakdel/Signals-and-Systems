الف)

```python
import matplotlib.pyplot as plt
import numpy as np

sample_rate = 500
t = np.arange(0, 2, 1/sample_rate)
freq = 3
phase = np.pi / 3
amplitude = 2

sine = amplitude * np.sin(2 * np.pi * freq * t + phase)
plt.plot(t, sine)
plt.title("Sine Wave")
plt.show()
```

ب)

```python
complex_sine = amplitude * np.exp(1j * 2 * np.pi * freq * t + phase)
plt.plot(t, np.real(complex_sine))
plt.plot(t, np.imag(complex_sine))
plt.legend(["real", "imag"], loc ="lower right")
plt.title("Complex Sine Wave")
plt.show()
```

ج)

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.gca(projection='3d')
ax.plot(t, np.real(complex_sine), np.imag(complex_sine))
plt.title("3D Wave of Complex Sine")
plt.xlabel("Time")
plt.ylabel("Real")
ax.set_zlabel("Imaginary")
plt.show()
```

```python
sample_rate = 1000
t = np.arange(-1, 1, 1/sample_rate)
amplitude = np.exp(-(t**2))
freq = 5
phase = 0
f_t = amplitude * np.sin(2 * np.pi * freq * t + phase)
plt.plot(t, f_t)
plt.title("f(t)")
plt.show()
```

2.

```python
def find_dots(freqs):
  y = []
  for f in freqs:
    new_sine = np.sin(2 * np.pi * f * t)
    y.append(np.dot(new_sine, f_t))
  return y

freqs = np.arange(2, 10, 0.5)
sines_dot_f = find_dots(freqs)

plt.stem(freqs, sines_dot_f)
plt.title("f(t) dot Sines with diff freqs")
plt.xlabel("Frequnecy")
plt.ylabel("Dot Output")
plt.show()
```

3.

```python
#theta = pi/2
phase = np.pi/2
f_t = amplitude * np.sin(2 * np.pi * freq * t + phase)
plt.plot(t, f_t)
plt.title("f(t) with phase pi/2")
plt.show()

sines_dot_f = find_dots(freqs)
plt.stem(freqs, sines_dot_f)
plt.title("f(t) with phase pi/2 dot Sines with diff freqs")
plt.show()

#theta = pi/4
phase = np.pi/4
f_t = amplitude * np.sin(2 * np.pi * freq * t + phase)
plt.plot(t, f_t)
plt.title("f(t) with phase pi/4")
plt.show()

sines_dot_f = find_dots(freqs)
plt.stem(freqs, sines_dot_f)
plt.title("f(t) with phase pi/4 dot Sines with diff freqs")
plt.show()
```

```python
amplitude = np.exp(-(t**2))
freq = 5
phase = 0
f_t = amplitude * np.sin(2 * np.pi * freq * t + phase)

def find_dots_complex(freqs, f_t):
  y = []
  for f in freqs:
    new_sine = np.exp(2 * np.pi * f * 1j * t)
    y.append(np.vdot(f_t, new_sine))
  return y

def stem_plots(freqs, dots_out, phase):
  fig=plt.figure(figsize=(10,5))
  plt.stem(freqs, dots_out)

  plt.title("f(t) with phase " + phase + " dot Complex Sines ")
  plt.xlabel("Frequnecy")
  plt.ylabel("Dot Output")
  
  plt.show()

dots_output = find_dots_complex(freqs, f_t)
size_dots_out = np.sqrt((np.real(dots_output))**2 + (np.imag(dots_output))**2)
stem_plots(freqs, size_dots_out, "0")

phase = np.pi/2
f_t = amplitude * np.sin(2 * np.pi * freq * t + phase)
dots_output = find_dots_complex(freqs, f_t)
size_dots_out = np.sqrt(np.real(dots_output)**2 + np.imag(dots_output)**2)
stem_plots(freqs, size_dots_out, "pi/2")

phase = np.pi/4
f_t = amplitude * np.sin(2 * np.pi * freq * t + phase)
dots_output = find_dots_complex(freqs, f_t)
size_dots_out = np.sqrt(np.real(dots_output)**2 + np.imag(dots_output)**2)
stem_plots(freqs, size_dots_out, "pi/4")
```

```python
t = np.linspace(-1, 1, 100)
phase = np.linspace(0, 7 * np.pi / 2, 100)

def dot_product(phase_, t_):
  out = []
  for ph in phase_:
    f_t = np.sin(2 * np.pi * 5 * t + ph) * np.exp(-(t**2))
    out.append(np.dot(f_t , np.sin(2 * np.pi * ph * t)))
  return out

plt.plot(t, dot_product(phase, t))
plt.title("Dot f(t) and Sine Waves")
plt.show()

def dot_product2(phase_, t_):
  out = []
  for ph in phase_:
    f_t = np.sin(2 * np.pi * 5 * t + ph) * np.exp(-(t**2))
    out.append(np.vdot(f_t , np.sin(1j*2 * np.pi * ph * t)))
  return out

plt.plot(np.real(dot_product2(phase, t)), np.imag(dot_product2(phase, t)))
plt.title("Dot f(t) and Complex Sine Waves")
plt.xlabel("Real")
plt.xlabel("Imaginary")
plt.show()
```
