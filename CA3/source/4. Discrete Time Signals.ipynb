# **Sequences**

```python
from lcapy import j, n, exp

x1 = exp((-1 * j * pi * n)/2)*u(n)
seq_x1 = x1.seq((0, 5))
seq_x1
```

```python
x2 = n*u(n)
seq_x2 = x2.seq((0, 5))
seq_x2
```

**ب)**

```python
seq_x3 = seq_x1 * 2
seq_x3.as_array()[5:11]
```

```python
seq_x3 = seq_x1 * 2
x4 = []
for m in range(0, 6):
  temp_x3 = seq_x3.as_array()[5-m:11-m]
  temp_x3 = temp_x3[0:m+1][::-1]
  temp_x2 = seq_x2.as_array()
  temp = []

  for i in range (0, m + 1):
    temp.append(temp_x3[i])

  temp_x3 = seq_x3.as_array()[5-m:11-m][::-1]
  for i in range(0, 6-(m+1)):
    temp.append(temp_x3[i])
    
  for i in range(0, 6):
    temp[i] = temp[i] * temp_x2[i]
  
  sum = 0
  for i in range(0, 6):
    sum += temp[i]
  x4.append(sum)
```

# **Discrete Transforms**

```python
from lcapy import n, exp
a = symbol('a')
x_n = ((0.5 - a)**n) * u(n) + 10 * ((a+(1/3))**n) * u(n)
x_n.expr
```

**الف)**

```python
from lcapy.discretetime import n, z

defs = {'a': 1}
x1 = x_n.subs(defs)
X1 = x1(z)
X1.plot(title = "Poles and Zeros For a = 1")

defs = {'a': 1.5}
x2 = x_n.subs(defs)
X2 = x2(z)
X2.plot(title = "Poles and Zeros For a = 1.5")

defs = {'a': 3}
x3 = x_n.subs(defs)
X3 = x3(z)
X3.plot(title = "Poles and Zeros For a = 3")
```

**ب)**

```python
from lcapy.discretetime import n, z

x = n * a**n * u(n)
x.expr
```
